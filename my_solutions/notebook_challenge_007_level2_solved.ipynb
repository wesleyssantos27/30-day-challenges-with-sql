````python
%load_ext sql
%sql sqlite://///tmp/github/30-day-challenges-with-sql/challenges/utils/db/database.db
````

````python
%%sql
--Identificação do conteudo da tabela Track
select * from track
limit 5
````

````python
%%sql
--Identificação do conteudo da tabela Album
select * from album
limit 5
````

````python
%%sql
--Identificação do conteudo da tabela Artist
select * from artist
limit 5
````

````python
%%sql
--Identificação do conteudo da tabela Genre
select * from Genre
limit 5
````

````python
%%sql
--Criação da tabela Track2 à partir do select com join entre as demais tabelas
create table if not exists track2
as
select 
  track.TrackId, 
  artist.name as artista, 
  track.name as musica, 
  track.composer as compositor, 
  genre.name as genero, 
  cast(track.bytes/1024 as int) as size_kb
from track
join album on track.AlbumId = album.AlbumId
join artist on album.artistid = artist.artistid
join genre on track.genreid = genre.genreid
````

````python
%%sql
--Exibição da nova tabela criada
select * from track2
````
